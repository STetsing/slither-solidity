```python
import numpy as np 
import pandas as pd
import os 
from datasets import Dataset
from slither.slither import Slither  
from slither_sol_helpers import *

# instal the sol version to be used. see https://github.com/crytic/solc-select

# Data path
data_dir = 'full_match'
```

```python
def gen_data():
    i = 0
    for pth, dr, fls in os.walk(data_dir):
        i += 1
        if i == 10:
            break
        if len(fls) > 0: 
            for f in fls:
                # with open(os.path.join(pth, f)) as f:
                #     lines = f.readlines()
                # print(os.path.join(pth, f))
                
                if '.sol' in f:
                    yield {'filename':os.path.join(pth, f), 'is_sol':True, 'is_json':False}
                elif '.json' in f: 
                    yield {'filename':os.path.join(pth, f), 'is_sol':False, 'is_json':True}
                else:
                    yield {'filename':os.path.join(pth, f), 'is_sol':False, 'is_json':False}


```

```python
def process_files():
    i = 0
    for pth, dr, fls in os.walk(data_dir):
        i += 1
        if i == 100:
            break

        # find the meta file
        if 'metadata.json' not in fls:
            continue
        
        # set the right solidity version 
        set_sol_version(detect_sol_version(os.path.join(pth, 'metadata.json')))

        # slither each sol file in the contrat diectory
        print('#'*100)
        for p, d, f in os.walk(pth):
            for x in f:
                if '.sol' in x:
                    res = slither_sol_file(os.path.join(p, x))
                    print('-'*100)
                    print(res)
                    print('-'*100)
                    print(os.path.join(p, x))

process_files()
```

```python
ds = Dataset.from_generator(gen_data)
```
